# Task 1: Exploring and Visualizing the Iris Dataset
**Objective:** Understand how to load, summarize, and visualize a dataset using Python.
**Libraries Used:** Pandas, Matplotlib, Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dataset load karne ka best tarika (taake GitHub par error na aaye)
df = sns.load_dataset('iris')
print("Dataset loaded successfully!")
```

## 1. Dataset Structure
Checking the shape, columns, and the first few rows of the data.

```python
# Displaying shape, columns and head
print("Shape of dataset:", df.shape)
print("Columns in dataset:", df.columns)
df.head()
```

## 2. Statistical Summary
Using `.describe()` to see the mean, standard deviation, and quartiles of the numerical features.

```python
df.describe()
```

## 3. Visualizations
### 3.1 Scatter Plot (Analyzing Relationships)
A scatter plot helps us see how 'Sepal Length' and 'Sepal Width' vary across different species.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', s=100)
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

### 3.2 Histogram (Data Distribution)
Checking the frequency distribution of 'Petal Length'.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['petal_length'], kde=True, color='blue')
plt.title('Distribution of Petal Length')
plt.show()
```

### 3.3 Box Plot (Outliers and Spread)
Analyzing the spread of 'Petal Width' and checking for any outliers.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='petal_width', data=df)
plt.title('Petal Width by Species')
plt.show()
```

## Conclusion
- The dataset is well-structured with 150 rows and 5 columns.
- Setosa is easily separable from the other two species.
- No major outliers were found in petal measurements.
